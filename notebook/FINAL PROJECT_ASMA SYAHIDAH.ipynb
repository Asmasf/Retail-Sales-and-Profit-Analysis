## cek outlier

```python
numeric_cols = [
    'Is Return',
    'Qty Sold',
    'Sales Amount',
    'Cogs',
    'Number of Transactions'
]

for col in numeric_cols:
    # Menghitung IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    # Menghitung jumlah outlier
    outlier_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    outlier_pct = (outlier_count / len(df)) * 100

    # Menampilkan hasil
    print("=" * 60)
    print(f"Kolom              : {col}")
    print(f"Q1                 : {Q1:.2f}")
    print(f"Q3                 : {Q3:.2f}")
    print(f"IQR                : {IQR:.2f}")
    print(f"Lower Bound        : {lower_bound:.2f}")
    print(f"Upper Bound        : {upper_bound:.2f}")
    print(f"Jumlah Outlier     : {outlier_count}")
    print(f"Persentase Outlier : {outlier_pct:.2f}%")

    # Visualisasi Boxplot
    plt.figure(figsize=(9, 2.5))

    plt.boxplot(
        df[col],
        vert=False,
        patch_artist=True,
        boxprops=dict(facecolor='skyblue'),
        medianprops=dict(color='red', linewidth=2),
        whiskerprops=dict(color='black'),
        capprops=dict(color='black'),
        flierprops=dict(
            marker='o',
            markersize=3,
            markerfacecolor='orange',
            alpha=0.5
        )
    )

    plt.title(f'Boxplot of {col}', fontsize=13, fontweight='bold')
    plt.xlabel(col)
    plt.grid(axis='x', linestyle='--', alpha=0.4)

    plt.show()
```

# Data Cleaning 

### mengubah tipe data 

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])

df.info()
```

### menghapus kolom Reason of Return

```python
df = df.drop(columns=['Reason of Return'])
```

```python
df.isnull().sum()
```

# Feature Engineering

## Profit

```python
df['Profit'] = df['Sales Amount'] - df['Cogs']
```

```python
print(df['Profit'].sum())
```

```python
(df['Profit'] == 0).sum()
```

## Profit Margin (%)

```python
import numpy as np

df['Profit Margin (%)'] = np.where(
    df['Sales Amount'] != 0,
    df['Profit'] / df['Sales Amount'],
    0
)
```

## Month

```python
df['Month'] = df['Transaction Date'].dt.month_name()
```

```python
month_order = [
    'January','February','March','April',
    'May','June','July','August',
    'September','October','November','December'
]

df['Month'] = pd.Categorical(
    df['Month'],
    categories=month_order,
    ordered=True
)
```

## Quarter

```python
df['Quarter'] = 'Q' + df['Transaction Date'].dt.quarter.astype(str)
```

## Year

```python
df['Year'] = df['Transaction Date'].dt.year
```

## Profit Status

```python
df['Profit Status'] = np.where(
    df['Profit'] >= 0,
    'Profit',
    'Loss'
)
```
